```python
input_files = ['edgar_allan_poe.txt', 'robert_frost.txt']
```

```python
# Now we need to store each line as a sample into a list

input =[]
labels =[]

for label, f in enumerate(input_files):
  print("The label of {} is {}".format(f,label))
  for line in open(f):
    line = line.rstrip().lower()
    if line:
      line.translate(str.maketrans('','', string.punctuation))
      input.append(line)
      labels.append(label)
```

```python
train_input, test_input, train_label, test_label = train_test_split(input, labels, test_size=0.3)
```

```python
vocabulary = {'unknown':0}
index =1

for line in train_input:
  words= line.split()
  for word in words:
    if word not in vocabulary:
      vocabulary[word]= index
      index+=1
```

```python
vocabulary
```

```python
 # Now we can convert the list containing each line as a sample into integers from the dictionary of word to index mapping
new_train_input=[]
new_test_input =[]

for line in train_input:
  words= line.split()
  line_int = [vocabulary[word] for word in words]
  new_train_input.append(line_int)

for line in test_input:
  line_int=[]
  words= line.split()
  for word in words:
    if word in vocabulary:
      line_int.append(vocabulary[word])
    else:
      line_int.append(vocabulary['unknown'])
  new_test_input.append(line_int)
```

```python
new_train_input[:5]
```

```python
v = len(vocabulary)
A0 = np.ones((v,v))    #Add One way smoothing
pi0 = np.ones(v)

A1 = np.ones((v,v))
pi1= np.ones(v)
```

```python
def computation(int_input, A, pi):
  for int_tokens in int_input:
    first_idx=None
    for idx in int_tokens:
      if first_idx is None:
         pi[idx] +=1
      else:
         A[first_idx,idx] +=1
         first_idx = idx
```

```python
computation([t for t,y in zip(new_train_input, train_label) if y==0], A0, pi0)
computation([t for t,y in zip(new_train_input, train_label) if y==1], A1, pi1)
```

```python
# Lets normalize A and pi so that we get that in terms of probabilities
A0 /= A0.sum(axis=1)
pi0/= pi0.sum()

A1 /= A1.sum(axis=1)
pi1/= pi1.sum()
```

```python
# Lets take the log of initial state matrix and state transition matrix since we are only interested in the trend
log_A0 = np.log(A0)
log_pi0 = np.log(pi0)

log_A1 = np.log(A1)
log_pi1 = np.log(pi1)
```

Lets now calculate prior probabilities,i.e, p(class=k)

```python
total_length = len(train_label)

total_e = sum(1 for y in train_label if y==0)
prior_e= total_e/total_length
log_prior0 = np.log(prior_e)
print(prior_e)

total_r = sum(1 for y in train_label if y==1)
prior_r= total_r/total_length
log_prior1 = np.log(prior_r)
print(prior_r)
```

```python
# Now we need to calculate the posterior
def likelihood(int_input, log_A, log_pi):
    log_likelihood=0
    for index in int_input:
        first_index= None
        if first_index is None:
          log_likelihood += log_pi[index]
          first_index= index
        else:
          log_likelihood += log_A[first_index,index]
          first_index= index
    return log_likelihood
```

```python
def predict(int_input):
  predictions= np.zeros(len(int_input))
  for i in range(len(int_input)):
      posterior_edgar= (likelihood(int_input[i], log_A0, log_pi0)) + log_prior0
      posterior_robert = (likelihood(int_input[i], log_A1, log_pi1)) + log_prior1
      if posterior_edgar> posterior_robert:
         predictions[i] = 0
      else:
         predictions[i] = 1


  return predictions
```

```python
p_train = predict(new_train_input)
print(np.mean(p_train==train_label))
```

```python
p_test = predict(new_test_input)
print(np.mean(p_test==test_label))
```
