```python
# Import Libraries
import numpy as np
import string
```

```python
f='robert_frost.txt'
tokens=[]

for line in open(f):
  if line:
    line = line.translate(str.maketrans('','', string.punctuation))
    line = line.rstrip().lower().split()
    tokens.append(line)
```

```python
tokens
```

```python
def modify_dict(dictionary, key, value):
  if key not in dictionary:
    dictionary[key]=[]
  dictionary[key].append(value)
  return dictionary
```

```python
initial_state={}
first_state={}
second_state={}

for t in tokens:
  len_token = len(t)
  for i in range(len_token):
    if i==0:                    # Corresponds to the 1st word in the line
       initial_state[t[i]] = initial_state.get(t[i],0.) + 1
    else:
      if i== len_token -1 :     # This corresponds to it being the last word of the sentence
        modify_dict(second_state, (t[i-1],t[i]), 'END')
      if i==1:                   #  This corresponds to the 2nd word in the line
        modify_dict(first_state, t[i-1], t[i])
      else:
        modify_dict(second_state, (t[i-2],t[i-1]), t[i])
```

```python
# Define a function for calculating the total of first_state and second_state matrices
class State_Total_Prob:
     def __init__(self, dictionary):
      self.dictionary= dictionary

     def calculate_total(self):
        state_total={}
        for key, value in self.dictionary.items():
          state_total[key] ={}
          for i in value:
            if i not in state_total[key]:
              state_total[key][i] =1
            else:
              state_total[key][i] +=1
        return state_total

     def prob(self):
        state_tot = self.calculate_total()
        state_prob={}

        for key,value in state_tot.items():
           total_sum=0
           for val in value.values():
               total_sum += val
           state_prob[key] ={}
           for k,val in value.items():
              state_prob[key][k] = val / total_sum

        return state_prob
```

```python
# Let's create objects
first_obj = State_Total_Prob(first_state)
second_obj = State_Total_Prob(second_state)

first_prob = first_obj.prob()
second_prob = second_obj.prob()
```

```python
total_initial=0.0
initial_prob={}
for key, value in initial_state.items():
  total_initial += value

for key, value in initial_state.items():
  initial_prob[key] = value/total_initial
```

```python
def sample_word(d):
  # print "d:", d
  p0 = np.random.random()
  # print "p0:", p0
  cumulative = 0
  for t, p in d.items():
    cumulative += p
    if p0 < cumulative:
      return t
  assert(False) # should never get here
```

```python
def generate():
  for i in range(4): # generate 4 lines
    sentence = []

    # initial word
    w0 = sample_word(initial_prob)
    sentence.append(w0)

    # sample second word
    w1 = sample_word(first_prob[w0])
    sentence.append(w1)

    # second-order transitions until END
    while True:
      w2 = sample_word(second_prob[(w0, w1)])
      if w2 == 'END':
        break
      sentence.append(w2)
      w0 = w1
      w1 = w2
    print(' '.join(sentence))
```

```python
generate()
```
